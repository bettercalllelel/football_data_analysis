```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import interpolate
from adjustText import adjust_text
```

```python
df = pd.read_csv('Atacantes.csv')
```

```python
x = df['Minutos']
y = df['Participacoes']
z = df['Nome']
df
```

```python
plt.figure(figsize=(20,10))
plt.scatter(x,y, color = 'green', marker = 'H', s = 200, edgecolor = 'red')
plt.grid(linestyle = '--')
plt.xlabel('Minutos em campo', fontsize = 15)
plt.ylabel('Participações em gols\n(Gols e assistências)', fontsize = 15)
plt.annotate('@bettercalllelel', (25,-2.5), fontsize = 15, annotation_clip=False)
plt.annotate('fonte: transfermarkt', (2025,-2.5), fontsize = 15, annotation_clip=False)
plt.title('Produtividade dos Atacantes do Fluminense em 2021\n(considerando Brasileirão, Libertadores e Copa do Brasil)', fontsize = 18)
texts = []
for i, txt in enumerate(z):
    texts.append(plt.annotate(txt, (x[i], y[i] + 0.4), xytext=(x[i],y[i]+.3) , color = 'red', fontsize = 15))

adjust_text(texts)
plt.savefig('fluminense.png', dpi = 600)
```
